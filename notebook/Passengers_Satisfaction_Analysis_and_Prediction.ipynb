<h6 style="font-family: 'Arial'; font-size: 20px; font-style: italic; color:rgb(125, 183, 255);">Analysis of Categorical Variables with Limited Unique Values Distribution, Value Counts and Percentage Distribution</h6>

```python
pd.set_option('display.max_colwidth', None)  # Disable column width truncation
```

```python
# List to store the analysis results for each column
analysis_results = []

# Function to analyze each column
def analyze_column(data, column_name):
    column_analysis = {}
    
    # Number of unique values in the column
    num_unique = data[column_name].nunique()
    column_analysis['Column Name'] = column_name
    column_analysis['Number of Unique Values'] = num_unique
    
    # Unique values in the column
    unique_values = data[column_name].unique()
    column_analysis['Unique Values'] = unique_values.tolist()
    
    # Value counts and percentage distribution
    value_counts = data[column_name].value_counts()
    percentage_distribution = (value_counts / len(data)) * 100
    
    # Format the percentages to 2 decimal points and add the '%' symbol
    formatted_percentage = percentage_distribution.apply(lambda x: f"{x:.2f}%")
    column_analysis['Value Counts'] = value_counts.to_dict()
    column_analysis['Percentage Distribution'] = formatted_percentage.to_dict()
    
    
    return column_analysis
```

```python
# Iterate over all columns in the dataframe to analyze them dynamically
for column in data.columns:
    if data[column].nunique() <= 10:  # Only process columns with 10 or fewer unique values
        result = analyze_column(data, column)
        analysis_results.append(result)
```

```python
# Convert the results into a DataFrame
analysis_df = pd.DataFrame(analysis_results)
analysis_df
```

Descriptions on when to use mean, median, or mode imputation:

| **Method**       | **Description**                                          | **When to Use**                                                                                     | **Pandas Implementation**                                        | **Scikit-learn Implementation**                        |
|------------------|----------------------------------------------------------|----------------------------------------------------------------------------------------------------|-----------------------------------------------------------------|--------------------------------------------------------|
| **Mean**         | Replaces missing values with the mean of the column.     | Use when data is normally distributed without significant outliers.                              | `df.fillna(df.mean(), inplace=True)`                             | `from sklearn.impute import SimpleImputer`<br>`imputer = SimpleImputer(strategy='mean')`<br>`df = imputer.fit_transform(df)` |
| **Median**       | Replaces missing values with the median of the column.   | Use when data contains outliers or is skewed; median is robust to outliers.                       | `df.fillna(df.median(), inplace=True)`                           | `from sklearn.impute import SimpleImputer`<br>`imputer = SimpleImputer(strategy='median')`<br>`df = imputer.fit_transform(df)`|
| **Mode**         | Replaces missing values with the mode of the column.     | Use when dealing with categorical data or when you want to impute based on the most common value. | `df.fillna(df.mode().iloc[0], inplace=True)`                     | `from sklearn.impute import SimpleImputer`<br>`imputer = SimpleImputer(strategy='most_frequent')`<br>`df = imputer.fit_transform(df)` |

### Notes:
- **Mean**: Best for continuous numerical data without extreme values.
- **Median**: Best for numerical data with outliers or skewed distributions.
- **Mode**: Best for categorical data or when the most frequent value is a logical choice for imputation.

```python
sns.set_theme(style="whitegrid")

# Compute the correlation matrix
corr_matrix = data[numerical_cols].corr()

# Mask the upper triangle to focus on the lower half of the correlation matrix
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))


plt.figure(figsize=(14, 8))
sns.heatmap(
    corr_matrix, 
    annot=True, 
    cmap='GnBu',
    fmt='.2f', 
    annot_kws={"size": 8, "weight": "bold", "family": "serif"},
    mask=mask  # Mask applied to display only the lower half of the correlation matrix
)

plt.title('Correlation Analysis of Airline Passenger Satisfaction Features', 
          fontsize=14, fontweight='bold', family='Georgia')

plt.xticks(rotation=90, ha='right', fontsize=8, family='Georgia') 
plt.yticks(rotation=0, fontsize=8, family='Georgia')
plt.tight_layout()
plt.show()
```

* ##### Top Correlations in Airline Passenger Satisfaction Features


**Key Points:**

1. **Positive Correlations Near +1**: Indicate a strong direct relationship, where both variables increase together.

2. **Negative Correlations Near -1**: Indicate a strong inverse relationship, where one variable increases as the other decreases.

3. **Sorting by Absolute Value**: Sorting correlations by their absolute value ensures both strong positive and negative relationships are included, highlighting the most significant feature interactions regardless of direction.

4. **Masking the Upper Triangle**: The upper triangle of the correlation matrix is masked to avoid redundant correlations (since `corr(A, B)` is the same as `corr(B, A)`).

5. **Excluding Self-Correlations**: Self-correlations (where a feature is correlated with itself) are removed to ensure the analysis focuses only on interactions between different features.



Note:

- **Lower Triangle Focus**: Masking the upper triangle highlights the lower half of the matrix, avoiding duplicate correlations and ensuring unique feature relationships are examined.
  
- **Clarification on Masking**: The upper triangle is masked because it mirrors the lower triangle, preventing repetition of correlation pairs.


```python
# Mask the upper triangle to focus on the lower half of the correlation matrix
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))  

# Apply the mask to the correlation matrix
triangular_corr = corr_matrix.where(mask)   #conditon of mask 

# Flatten the  triangular matrix and reset the index
corr_pairs = triangular_corr.stack().reset_index()

corr_pairs.columns = ['Column_1', 'Column_2', 'Correlation_Coefficient']

# Drop self-correlations (correlation of a column with itself)
corr_pairs = corr_pairs[corr_pairs['Column_1'] != corr_pairs['Column_2']]

# Sort by absolute correlation values to show the strongest correlations
corr_pairs_sorted = corr_pairs.loc[
    corr_pairs['Correlation_Coefficient'].abs().sort_values(ascending=False).index
]

# Show the top 10 correlations
top_correlations = corr_pairs_sorted.head(10).reset_index(drop=True)
top_correlations = pd.DataFrame(top_correlations)
top_correlations


```
